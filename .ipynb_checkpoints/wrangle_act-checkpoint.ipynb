### 3. Twitter API

[<a href="#Table-of-Contents">Back to Top</a>]

```python
consumer_key =''
consumer_secret =''
access_token =''
access_secret =''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)

tweet = api.get_status(666020888022790149, trim_user=False, include_my_retweet=False, include_entities=False, include_ext_alt_text=False, include_card_uri=False)
print(twit_archive['tweet_id'][0:5])
print(tweet.retweet_count)
print(tweet.favorite_count)
print(tweet)
```

```python
start=time.time()
tweet_list=[]
missing_tweets_id=[]
#loop

for tweet_id in twit_archive['tweet_id']:
    try:
        tweet_list.append(api.get_status(tweet_id, trim_user=False, include_my_retweet=False, include_entities=False, include_ext_alt_text=False, include_card_uri=False))
    except Exception as e:
        missing_tweets_id.append(tweet_id)
        
print(time.time()-start,'seconds')
```

```python
print(len(tweet_list))
print(len(missing_tweets_id))
print(tweet_list[0]._json)
```

```python
def to_json(item):
    return item._json

with open('tweet_json.txt', mode='w') as file :
        file.write(json.dumps(list(map(to_json, tweet_list)), indent=4))
```

```python
# This function just take needed columns 
def assess_json(tweet_json) :
    return {
        'tweet_id': str(tweet_json['id']),
        'retweet_count':int(tweet_json['retweet_count']),
        'favorite_count': int(tweet_json['favorite_count']),
        'followers_count': int(tweet_json['favorite_count'])
    }
def assess_json_header() :
    return ('tweet_id','retweet_count','favorite_count', 'followers_count')
```

```python
# download the columns (tweet_id,retweet_id and followers_count) in the list
df_list = []
with open ('tweet_json.txt', mode='r') as file :
    tweet_json_array = json.load(file) 
    for tweet_json in tweet_json_array:
        df_list.append(assess_json(tweet_json))
        
```

```python
# Create DataFrame from the list 
tweet_json=pd.DataFrame(df_list,columns= assess_json_header())
tweet_json.head()
```

```python
#Save the file as tweet_json.csv
tweet_json.to_csv('tweet_json.csv',index=False)
#read the tweet_json file 
tweet_json=pd.read_csv('tweet_json.csv')
```

##### <h3 style="color:IndianRed"> 4.Total number of tweet over the months</h3> 

```python
twitter_archive_master.dtypes
```

```python
#convert timestamp column to datetime type
twitter_archive_master['timestamp']=pd.to_datetime(twitter_archive_master['timestamp'],utc=False)
```

```python
tweets_time=twitter_archive_master.groupby(twitter_archive_master.timestamp.dt.to_period('m').rename('date'))['tweet_id'].count()
tweets_time=tweets_time.to_frame().reset_index()
tweets_time.sort_values(by='tweet_id')
#.rename('tweet_id':'number_of_tweets')
```

```python
ax=sns.lineplot(x=tweets_time['date'].astype(str),y=tweets_time['tweet_id'])
ax.tick_params(axis='x',labelrotation=90)
ax.set(xlabel='date',ylabel='Number of tweets')
plt.title('Number of tweets over time')
plt.show()
```

> `We can see that the 4 first month of WeRateDogs account creation was the period when the account made the most tweets in a month with a pic in the month of December 2015.`

>`We can see that the number of tweets drops at February 2016 and start to differ a little bit since March 2016. It might be because the account is now quite popular and doesn't necessarily need to post a lot. it might also have another explanation.`
